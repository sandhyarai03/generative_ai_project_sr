```python
def is_pali(string : str)-> bool:
    newStr = string.lower()
    return newStr == newStr[::-1]
print(is_pali("Racecars"))


```

```python

def missingNum(numList, MaxNum):
    expectedNum = MaxNum * (MaxNum + 1) //2
    return expectedNum - sum(numList)


NumList = [1,2,3,4,5,7]
print(missingNum (NumList, 7))
```

```python
def countOfNu(n):
    numCount = sum(n)
    return numCount

lis = [1,2,3,4,5]
print(countOfNu(lis))
```

```python
from collections import Counter
def wordCount(Sen):
    count = Counter(Sen.split())
    return count
text= 'sdf dsa sdf rfd vfr vfr cde vfr'
print(wordCount(text))
```
