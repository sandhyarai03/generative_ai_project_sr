```python
from openai import OpenAI
import os
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path="../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=my_api_key)
#print(my_api_key)

def ask_movie(userPrompt: str) -> str:

    response = client.chat.completions.create(
        model="gpt-5-nano", #"gpt-5-turbo", #"gpt-5", #"gpt-5-mini", #"gpt-5-nano"
        messages=[
        {"role": "system", "content": "You are a helpful assistant. Answer the user's questions concisely."},
        {"role": "user", "content": userPrompt}
            # {"role": "user", "content": f"Here is the context: \n\ {context} \n\. \n\n Now, answer the question: {prompt}")
        ]
    )

    print (response)
    print (response.choices[0].message.content)


userInput = input("how may i help you")

while True:

    ask_movie(userInput) 
    userInput = input("how may i help you again ")
    if userInput == "qw" :
        print ("Thank You for choosing AI")
        break
    




```
